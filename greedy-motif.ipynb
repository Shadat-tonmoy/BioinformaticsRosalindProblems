```python

def get_id(ch):
    if ch == 'A' :
        return 0
    if ch == 'C' :
        return 1
    if ch == 'G' :
        return 2
    if ch == 'T' :
        return 3
    

def get_score(str):
    t=len(str)
    k=len(str[0])
    
    ret=0
    
    for i in range(k):
        cnt=[0]*4
        for j in range(t):
            v = get_id(str[j][i])
            cnt[v]+=1
        mx=0  
        for j in range(4):
            if cnt[j]>mx:
                mx=cnt[j]
        ret+=t-mx
    return ret

def get_profile(str):
    t=len(str)
    k=len(str[0])
    
    ret=[]
    for i in range(k):
        ret.append([0.0]*4)
    for i in range(k):
        cnt=[0]*4
        for j in range(t):
            v = get_id(str[j][i])
            cnt[v]+=1 
        for j in range(4):
            ret[i][j] = cnt[j]/(0.0+t)
    return ret
    
    
def get_profile_most(profile,str) :
    k=len(profile)
    mx_prob=0.0
    ret=str[0:k]
    for i in range(len(str)-k+1):
        ss=str[i:i+k]
        cur_prob=1.0
        for j in range(k):
            v=get_id(str[i+j])
            cur_prob*=profile[j][v]
        if cur_prob > mx_prob:
            mx_prob=cur_prob
            ret=ss
    return ret
    


def greedy_motif(str,k):
    best=[]
    for i in range(len(str)):
        ss=str[i][0:k]
        best.append(ss)
    #print(best)
    mn_score=get_score(best)
    print(best," ",mn_score)
    for i in range(len(str[0])-k+1):
        ss=str[0][i:i+k]
        now=[]
        now.append(ss)
        for j in range(1,len(str)):
            profile=get_profile(now)
            ss=get_profile_most(profile,str[j])
          #  print(profile," ",str[j]," ",ss)
            now.append(ss)
        score = get_score(now)
        print(now," ",score)
        if score < mn_score:
            mn_score=score
            best=now
       # break
    
    return best
    
    

k=3
t=5

str=[]

str.append("GGCGTTCAGGCA")
str.append("AAGAATCAGTCA")
str.append("CAAGGAGTTCGC")
str.append("CACGTCAATCAC")
str.append("CAATAATATTCG")


#print(get_score(str))
#print(get_profile(str))
#print(get_profile_most(get_profile(str),str[0]))

print(greedy_motif(str,3))







```
